```python
import string
movie_description_tokens=[]
movie_summaries_tokens=[]
def clean(text):
    text = text.lower()
    printable = set(string.printable)
    return filter(lambda x: x in printable, text)

for i in range(2458):
    cleanDescription=clean(movie_description[i])
    cleanSummary = clean(target_summary[i])
    movie_summaries_tokens.append(nltk.word_tokenize(str(cleanSummary)))
    movie_description_tokens.append(nltk.word_tokenize(str(cleanDescription)))
    
#print len(movie_description_tokens)    
```

```python
filename = 'glove.6B/glove.6B.50d.txt'
def loadGloVe(filename):
    glovevocab = []
    gloveembd = []
    file = open(filename,'r')
    for line in file.readlines():
        row = line.strip().split(' ')
        glovevocab.append(row[0])
        gloveembd.append(row[1:])
    print('Loaded GloVe!')
    file.close()
    return glovevocab,gloveembd
glovevocab,gloveembd = loadGloVe(filename)

embedding = np.asarray(gloveembd)
embedding = embedding.astype(np.float32)

word_vec_dim = len(embedding[0])
```

```python
def np_nearest_neighbour(x):
    #returns array in embedding that's most similar (in terms of cosine similarity) to x
        
    xdoty = np.multiply(embedding,x)
    xdoty = np.sum(xdoty,1)
    xlen = np.square(x)
    xlen = np.sum(xlen,0)
    xlen = np.sqrt(xlen)
    ylen = np.square(embedding)
    ylen = np.sum(ylen,1)
    ylen = np.sqrt(ylen)
    xlenylen = np.multiply(xlen,ylen)
    cosine_similarities = np.divide(xdoty,xlenylen)

    return embedding[np.argmax(cosine_similarities)]
    


def word2vec(word):  # converts a given word into its vector representation
    if word in glovevocab:
        return embedding[glovevocab.index(word)]
    else:
        return embedding[glovevocab.index('unk')]

def vec2word(vec):   # converts a given vector representation into the represented word 
    for x in xrange(0, len(embedding)):
            if np.array_equal(embedding[x],np.asarray(vec)):
                return vocab[x]
    return vec2word(np_nearest_neighbour(np.asarray(vec)))
```
